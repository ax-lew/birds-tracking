```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
matrix_res_pd = pd.read_csv('matrix_coocurrence.csv', index_col=0)
```

```python

FEMALES = [11, 17, 20, 24, 26, 28, 30, 33, 34]
MALES = [10, 14, 21, 22, 23, 25, 27, 29, 31, 32]
```

```python
matrix_res_pd
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sn.heatmap(matrix_res_pd, annot=True, fmt='g', xticklabels=True, yticklabels=True, vmin=0, cbar_kws={'label': 'Coocurrencia'},ax=ax)
ax.set(xlabel='Macho', ylabel='Hembra')


fig.savefig('MatrizCoocurrencia.png')
```

```python
fig, ax = plt.subplots(figsize=(10,5))

res_pd = (matrix_res_pd/matrix_res_pd.sum()*100).round(2)
res_pd = res_pd.fillna(0)
    
sn.heatmap(res_pd, annot=True, fmt='g', xticklabels=True, yticklabels=True, vmin=0, vmax=100,cbar_kws={'label': 'Coocurrencia'},ax=ax)
ax.set(xlabel='Macho', ylabel='Hembra')


fig.savefig('MatrizCoocurrenciaRelativaMachos.png')
```
